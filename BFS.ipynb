```python
# Breadth first search 

# from collection import deque  
# deque() algorithms used ti add and remove from the list 

def bfs(graph , start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    
    
    while queue: 
        vertex = queue.popleft()
        for neighbor in graph[vertex]:
            if neighbor not in visited: 
                visited.add(neighbor)
                queue.append(neighbor)
    
    return visited 
```

```python
# Graph representation 
# We have the following graph 

from collections import deque

def bfs(graph, start):
    visited = set()  # Set to keep track of visited nodes.
    queue = deque([start])  # Initialize the queue with the starting node 
    bfs_order = []  # To store the order of visited nodes.
    
    visited.add(start)
    
    
    while queue: 
        # Dequeue operation 
        node = queue.popleft()  # Dequeue the next node  removes the first element of the list
        bfs_order.append(node)  # Add the node to the BFS order list 
        # The node is added to the bfs_order list
        
        # Enqueue all adjacent , unvisisted nodes.
        
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    
    
    return bfs_order 


# Example usage :

graph  = { 
          1: [2 , 3],
          2: [1, 4 , 5], 
          3: [1 , 6],
          4: [2],
          5: [2],
          6: [3]
          
          }

start_node = 1 
bfs_result = bfs(graph , start_node)
print("BFS Order:", bfs_result)
```

```python
# Python code Example 


def max_sum_subarray(arr , k):
    
    # Find the sum of the first window 
    max_sum  = sum(arr[:k])
    window_sum = max_sum 
    
    
    # Slide the window from start to end 
    for i in range(1 , len(arr) - k + 1): # (1 , 4)
        window_sum = window_sum - arr[i - 1] + arr[i + k - 1] #  For shrink and add
        max_sum = max(max_sum , window_sum)
    
    return max_sum



# Example usage:
arr = [2 , 1 , 5 , 1, 3 , 2]
k = 3
print(max_sum_subarray(arr, k))
```

```python
# Example 1 : longestr substring without repeating characters 


def length_of_length_substring(s : str) -> int:
    char_set = set()  # To keep track of charaters in the current window 
    left  = 0  # Left pointer of the sliding window 
    max_length  = 0 # Result for the maximum length of substring 
    
    for right in range(len(s)):
        
        # IF the charatcter at right pointer in set, remove characters from left 
        while s[right] in char_set: 
            char_set.remove(s[left])
            left += 1 
            
        
        # Add the character at right pointer to the set 
        char_set.add(s[right])
        
        
        # Update the max length 
        max_length = max(max_length , right - left + 1)
        
    return max_length

# Example usage 
s = "abcabcbb"
print(length_of_length_substring(s)) # Output: 3('abc')
```

```python
# Example 2 : Minimum size subarry Sum 
# Sliding window approach 

def min_subarry_len(target: int , nums: list) -> int:
    
    left = 0
    current_sum = 0 
    min_length = float('inf')
    
    
    for right in range(len(nums)):
        current_sum += nums[right]
        
        
        # Shrink the window untill the sum less than target
        while current_sum >= target: 
            min_length  = min(min_length , right - left + 1)
            current_sum -= nums[left]
            left += 1 


    return min_length if min_length != float('inf') else 0 



# Example usage 
nums = [2 , 3 , 1 , 2 , 4 , 3]
target  = 7 
print(min_subarry_len(target , nums))
```

```python
# Example 3 : Maximum Number of vowels in a substring of Given length


def max_vowels(s: str , k: int) -> int: 
    vowels  = set('aeiou')
    max_vowels_count = 0
    current_vowels_count = 0 
    
    
    # First window of size k 
    for i in range(k):
        if s[i] in vowels: 
            current_vowels_count += 1 
    
    
    max_vowels_count = current_vowels_count
    
    
    # Sliding the window 
    for i in range[k , len(s)]:
        if s[i] in vowels: 
            current_vowels_count += 1
        if s[i - k] in vowels: 
            current_vowels_count -= 1
            
        max_vowels_count  = max(max_vowels_count , current_vowels_count)
    return max_vowels_count

# Example usage: 
s = 'abciiidef'
k = 3
print(max_vowels(s , k))  # Output 3 ('iii)
```

```python
class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.graph_dict = {}
        for start, end in edges:
            if start in self.graph_dict:
                self.graph_dict[start].append(end)
            else:
                self.graph_dict[start] = [end]
        print("Graph Dict:", self.graph_dict)

    def get_paths(self, start, end, path=[]):
        path = path + [start]

        if start == end:
            return [path]

        if start not in self.graph_dict:
            return []

        paths = []
        for node in self.graph_dict[start]:
            if node not in path:
                new_paths = self.get_paths(node, end, path)
                for p in new_paths:
                    paths.append(p)

        return paths

    def get_shortest_path(self, start, end, path=[]):
        path = path + [start]

        if start == end:
            return path

        if start not in self.graph_dict:
            return None

        shortest_path = None
        for node in self.graph_dict[start]:
            if node not in path:
                sp = self.get_shortest_path(node, end, path)
                if sp:
                    if shortest_path is None or len(sp) < len(shortest_path):
                        shortest_path = sp

        return shortest_path

if __name__ == '__main__':

    routes = [
        ("Mumbai","Pune"),
        ("Mumbai", "Surat"),
        ("Surat", "Bangaluru"),
        ("Pune","Hyderabad"),
        ("Pune","Mysuru"),
        ("Hyderabad","Bangaluru"),
        ("Hyderabad", "Chennai"),
        ("Mysuru", "Bangaluru"),
        ("Chennai", "Bangaluru")
    ]

    routes = [
        ("Mumbai", "Paris"),
        ("Mumbai", "Dubai"),
        ("Paris", "Dubai"),
        ("Paris", "New York"),
        ("Dubai", "New York"),
        ("New York", "Toronto"),
    ]

    route_graph = Graph(routes)

    start = "Mumbai"
    end = "New York"

    print(f"All paths between: {start} and {end}: ",route_graph.get_paths(start,end))
    print(f"Shortest path between {start} and {end}: ", route_graph.get_shortest_path(start,end))

    start = "Dubai"
    end = "New York"

    print(f"All paths between: {start} and {end}: ",route_graph.get_paths(start,end))
    print(f"Shortest path between {start} and {end}: ", route_graph.get_shortest_path(start,end))
```
